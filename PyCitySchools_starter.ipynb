```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools=school_data['school_name'].count()
total_schools
```

```python
# Calculate the total number of students
total_number_students=int(student_data['Student ID'].count())
total_number_students
```

```python
# Calculate the total budget
total_budget=school_data['budget'].sum()
total_budget
```

```python
# Calculate the average math score
average_math_score=round(student_data['math_score'].mean(),6)
average_math_score
```

```python
# Calculate the average reading score
average_reading_score=round(student_data['reading_score'].mean(),6)
average_reading_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
pass_students_m=student_data.loc[student_data['math_score']>=70,:]
pass_math=pass_students_m.count()['student_name']
pass_math
pass_math_rate=round(((pass_math/total_number_students)*100),6)
pass_math_rate
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
pass_students_r=student_data.loc[student_data['reading_score']>=70,:]
pass_reading=pass_students_r.count()['Student ID']
pass_reading
pass_reading_rate=round(((pass_reading/total_number_students)*100),6)
pass_reading_rate
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
pass_students_b=student_data.loc[(student_data['reading_score']>=70)&(student_data['math_score']>=70),:]
pass_students_both=pass_students_b.count()['grade']
pass_students_both
pass_students_both_rate=round(((pass_students_both/total_number_students)*100),8)
pass_students_both_rate
```

```python
Distict_Summary =pd.DataFrame({'Total Schools':[total_schools],'Total Students':[total_number_students],'Total Budget':[total_budget],'Average Math Score':[average_math_score],'Average Reading Score':[average_reading_score],'%Pass Math':[pass_math_rate],'%Pass Reading':[pass_reading_rate],'%Overall Passing':[pass_students_both_rate]})
Distict_Summary['Total Students']=Distict_Summary['Total Students'].map("{:,}".format)
Distict_Summary['Total Budget']=Distict_Summary['Total Budget'].map("${:,.2f}".format)
Distict_Summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarizes key metrics about each school, including:

# School Name
df=school_data_complete.set_index('school_name')
df.head()

```

```python
# School Type
df_group=df.groupby(['school_name'])
df_group.head()
```

```python
school_type=df_group['type'].last()
school_type
```

```python
# Total Students
total_students=df_group['size'].count()
total_students
```

```python
# Total School Budget
total_school_budget=df_group['budget'].first()
total_school_budget
```

```python
# Per Student Budget
per_student_budget=total_school_budget/total_students
per_student_budget
```

```python
# Average Math Score
avergea_math_score=df_group['math_score'].mean()
avergea_math_score
```

```python
# Average Reading Score
avergea_reading_score=df_group['reading_score'].mean()
avergea_reading_score
```

```python
# % Passing Math
all_school_pass_math=df.loc[df['math_score']>=70,:]

all_school_passm=all_school_pass_math.count()

each_school_passm=all_school_pass_math.groupby(['school_name']).count()
each_school_passm
# each_school_passm

# each school pass rate= each school pass students/school total students
rate_math=each_school_passm['math_score']/total_students*100
rate_math
```

```python
# % Passing Readings
all_school_pass_readings=df.loc[df['reading_score']>=70,:]

all_school_passr=all_school_pass_readings.count()

each_school_passr=all_school_pass_readings.groupby(['school_name']).count()
each_school_passr
# # each_school_passr

# # each school pass rate= each school pass students/school total students
rate_reading=each_school_passr['reading_score']/total_students*100
rate_reading
```

```python
# % Passing overall
all_school_pass_overall=df.loc[(df['reading_score']>=70)&(df['math_score']>=70),:]

all_school_passo=all_school_pass_overall.count()

each_school_passo=all_school_pass_overall.groupby(['school_name']).count()
each_school_passo
# # # each_school_passr

# # # each school pass rate= each school pass students/school total students
rate_overall=each_school_passo['reading_score']/total_students*100
rate_overall
```

```python
new_df=pd.DataFrame({"type":school_type,"Total Students":total_students,"Total School Budget":total_school_budget,"Per Student Budget":per_student_budget,"Average Math Score":avergea_math_score,"Average Reading Score":avergea_reading_score,"% Passing Math":rate_math,"% Passing Reading":rate_reading,"% Overall Passing":rate_overall})
```
